```python
import requests
from bs4 import BeautifulSoup as bs
import re
import os
```

```python
data = ['https://en.wikipedia.org/wiki/Iron_Man'
'https://en.wikipedia.org/wiki/Thor_(Marvel_Cinematic_Universe)',
'https://en.wikipedia.org/wiki/Captain_America',
'https://en.wikipedia.org/wiki/Hulk']
```

```python
data
```

```python
url = 'https://en.wikipedia.org/wiki/Iron_Man'
response = requests.get(url)
html = response.content
soup = bs(html,'xml')
```

```python
soup
```

```python
all_h2=soup.find_all('h2')
for h2 in all_h2:
  print(h2.get_text(strip=True))
```

```python
all_p=soup.find_all('p')
for p in all_p:
  print(p.get_text(strip=True))
```

```python
all_p=soup.find_all('p')
for p in all_p:
  print(p.get_text(strip=True,separator = ''))
```

```python
all_p = soup.find_all('p')
for p in all_p:
  print(len(p.get_text(strip=True, separator='').split()))
```

```python
full_content_length = 0
all_p = soup.find_all('p')
for p in all_p:
  pragraph_length=(len(p.get_text(strip=True, separator='').split()))
  full_content_length = full_content_length + pragraph_length
```
